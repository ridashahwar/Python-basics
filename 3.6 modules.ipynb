### Modules 
Modules refer to a file containing Python statements and definations.

A file containing Python code, For ex: abc.py is called a module and its module name would be 'abc'.

We use modules to break down large programs into small manageable and organised files. Furthermore, modules provide reusability of code.

we can define our most used functions in a module and import it,instead of copying their definations into different programs.

### How to import a module?
we use import keyword to do this.





Using the module name we can access the function using dot(.) opearation.

### Python has a lot of standard modules available.
