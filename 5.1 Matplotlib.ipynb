### Pyplot Tutorial
Matplotlib.pyplot is a collection style functions that make matplotlib work like MATLAB.

Each pyplot functiokn makes some change to a figure ex: creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with lables, etc.

Matplotlib is an amazing visualization library in Python for 2D plots of arrays. Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the broader SciPy stack. It was introduced by John Hunter in the year 2002.

One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc
    

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2,4,6,4])
plt.ylabel('numbers')
plt.xlabel('indices')
plt.title('Myplot')
plt.show()
```

if you provide a single list or array to the plot() command, matplotlib assumes it is a sequence of y values, and automatically generates the x values for you. since python ranges start with 0 the plot x values y 

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid()
plt.show()
```

for every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot.

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.grid()

plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2)

# blue dashes, red squares and green triangles
plt.plot(t,t**2, 'b--', label='^2')
plt.plot(t,t**2.2, 'rs', label='^2.2')
plt.plot(t,t**2, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
x= [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y,linewidth= 7.0)
plt.grid()
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)

# use keyword args
plt.setp(lines[0],color = 'r',linewidth = 2.0)

#or matlab style string value pairs
plt.setp(lines[1],'color','g','linewidth',2.0)

plt.grid()
```

```python
def f(t):
     return np.exp(-t)*np.cos(2*np.pi*t)
    
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1)

plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1), 'b-')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)    # first figure
plt.subplot(211)  #the first subplot in the first figure
plt.plot([1,2,3])
plt.subplot(212)   #the second subplot in the first figure
plt.plot([4,5,6])

plt.figure(2)       #a second figure
plt.plot([4,5,6])   # creates a subplot(111) by default

plt.figure(1)        #figure1 current subplot(212) still current
plt.subplot(211)      #make subplot(211) in figure1 current
plt.title('easy as 1,2,3') #subplot 211 title
plt.show()
```
